```python
from scipy import stats
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# Распределения случайных величин в scipy

### **Новый интерфейс (1.15+)**

**Общая логика:** Все распределения - классы в модуле stats, записанные в **CamelCase**. Чтобы воспользоваться создаём экземпляр класса и используем его методы, чтобы получить желаемый результат.

Основные функции:
* **sample** - cгенерировать случайные значения
* **pdf** - Probability density function (плотность вероятности)
* **cdf** - Cumulative distribution function (значение функции распределения)
* **icdf** - Синоним: **ppf**. Inverse of cdf (значение случайной величины по заданной вероятности)
* **plot** - строит график плотности вероятности 

**Warning** в версии scipy 1.15 только нормальное и равномерное распределение могут быть созданы указанным выше способом, для остальных непрерывных распределений нужно пользоваться функцией [make_distribution](https://scipy.github.io/devdocs/reference/generated/scipy.stats.make_distribution.html)

```python
# Пример
X = stats.Normal(mu=3, sigma=4)
X.plot()
print("Вероятность того, что X меньше 2:", X.cdf(2))
print("С веротяностью 80% значение X меньше:", X.icdf(0.8))
```

### Старый интерфейс

**Общая логика** Отедльные классы представляют собой семейства распределений, а их экзмепляры - конкретные случайные величины (т.е. `stats.norm` - семейство всех нормальных распределений, а `stats.norm(loc=0, scale=1)` - стандартное нормальное распределение). При этом методы в первом и втором случае практически полностью совпадают, но в первом случае для каждого метода при вызове нужно конкретно указывать параметры распределения (т.е. `stats.norm.rvs(loc=1, scale=2)` и `stats.norm(loc=1, scale=2).rvs()` эквивалентны)

**Основные методы**
* **rvs** - сгенерировать случайные значения
* **pdf** - probability density function (плотность распределения)
* **cdf** - cumulative distribution functions (функция распределения)
* **sf**  - survival function (1-cdf)
* **ppf** - percent point function (обратная функция к cdf, значение случайной величины для заданной вероятности)
* **isf** - inverse survival function (обратная функция к sf)
* **stats** - mean, variance, skew, kurtosis
* **moments** - нецентральные моменты
